```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = "iris.csv"
df = pd.read_csv(iris)
print(df)
```

```python
print(df.info())
```

```python
print(df.isna().sum())
```

```python
correlation_matrix = np.corrcoef(df['petal_length'], df['petal_width'])
print(correlation_matrix)
```

```python
df['petal_ratio'] = df['petal_length'] / df['petal_width']
print(df.head())
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='viridis')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, hue='class', vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'petal_ratio'])
plt.show()
```

```python
sns.boxplot(data=df, x='class', y='petal_ratio')
plt.title('Petal Ratio by Species')
plt.show()
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_ratio', hue='class')
plt.title('Petal Length vs. Petal Width')
plt.show()
```
