```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv("reviews.csv")
print("--- Text ---")
print(df)
```

```python
if "ReviewText" in df.columns:
    print("Column 'ReviewText' exists.")
else:
    print("Column 'ReviewText' does not exist")
```

```python
df = df[df['ReviewText'].notna()]
print(df)
```

```python
df["TextLenght"] = df["ReviewText"].str.len()
print(df)
```

```python
df["WordCount"] = df["ReviewText"].str.split().str.len()
print(df)
```

```python
text = "".join(df["ReviewText"]).lower()
for ch in [".", ",", "!", "?", ":", ";"]:
    text = text.replace(ch, " ")
tokens = text.split()
print(tokens)
```

```python
top10 = Counter(tokens).most_common(10)
top10_df = pd.DataFrame(top10, columns=["Word", "Count"])
print("--- Top10 ---")
print(top10_df)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(top10_df["Word"],top10_df["Count"], color='skyblue')
plt.title("Top 10 Most Frequent Words in Reviews")
plt.xlabel("Word")
plt.ylabel("Count")
plt.show()
```
